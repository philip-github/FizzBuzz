```python
def fizzbuzz(number):
    
    lst = []
    for num in range(1,number+1):
        lst.append(num)
    my_list = lst
    base_3 = 3
    base_5 = 5
    num_3 = 3
    num_5 = 5
    m_3_base = 2
    m_5_base = 2
    for ele in my_list:
        
        if base_3 == base_5:
            my_list[base_3-1]= "FizzBuzz"
            base_3 = num_3 * m_3_base
            m_3_base +=1
            base_5 = num_5 * m_5_base
            m_5_base +=1
            
        elif ele == base_3:
            my_list[ele-1] = 'Fizz'
            base_3 = num_3 * m_3_base
            m_3_base += 1
            
        elif ele == base_5:
            my_list[ele-1] = 'Buzz'
            base_5 = num_5 * m_5_base
            m_5_base += 1
    return my_list

fizzbuzz(100) 
```
